# taking longtime to convert to unixtime

```python
import time
import datetime

longTime=test.loc[200:360:1,['longTime']]
longTime=longTime.as_matrix()
# print(type(longTime))
a=[]
date_time_array=[]
for k in longTime:
    a=np.append(a,k)

# print(a[0:])
# print(a[-2:])
count=0
for b in a[::-1]:
    b=int(b) # make plain integer
    str_b=str(b)
    c=str_b[-3:]
    new_str_b=str_b.replace(c, '',1)
    new_str_b_time=int(new_str_b)
    
    now_time=datetime.datetime.fromtimestamp(new_str_b_time)
    convert_time=now_time.strftime('%Y-%m-%d %H:%M:%S')
    
#     print('count_num: ',count,' unix_time: ',convert_time)
    date_time_obj = datetime.datetime.strptime(convert_time, '%Y-%m-%d %H:%M:%S')
    date_time_array=np.append(date_time_array,date_time_obj)
#     print('count_num: ',count,' convert_time: ',convert_time," date_time_obj: ",date_time_obj," str_time",new_str_b_time)
#     count+=1

# print(date_time_array.size)
for q in date_time_array[::1]:
# #     q=q.split(" ")
# #     split_q=np.append(split_q,q)
    print("count: ",count," date: ",q.date()," time: ",q.time())
    count+=1
# #     print(type(q))
```

# taken from stackoverflow

```python
#taken from stackoverflow

import matplotlib.pyplot as plt
import matplotlib.dates as md
import numpy as np
import datetime as dt
import time
%matplotlib inline

n=20 #20 times it will add duration with the local time
duration=1000
now=time.mktime(time.localtime()) #make timestamp
# print(now)
timestamps=np.linspace(now,now+duration,n)

for i in timestamps:
#     print(i)
    i_i=datetime.datetime.fromtimestamp(i)
    convert_time=i_i.strftime('%Y-%m-%d %H:%M:%S')
#     print(convert_time)
#     print('count_num: ',count,' unix_time: ',convert_time)
    date_time_obj = datetime.datetime.strptime(convert_time, '%Y-%m-%d %H:%M:%S')
#     date_time_array=np.append(date_time_array,date_time_obj)
#     print(date_time_obj)
#     print('\n')


# print(type(timestamps))
# print(timestamps)
dates=[dt.datetime.fromtimestamp(ts) for ts in timestamps]
# for t in dates:
#     print(t)
# print(type(dates))
# print(dates)

datenums=md.date2num(dates)
# for e in datenums:
#     print(e)
# print(type(datenums))
# print(datenums)

values=[]
for i in range (1,21,1):
    if i<10:
        i=i*i
    else:
        i=i+10
#     print(i)
    values=np.append(values,i)
    
# print(values)   


# values=np.sin((timestamps-now)/duration*2*np.pi)
plt.subplots_adjust(bottom=0.2)
plt.xticks( rotation=25 )
ax=plt.gca()
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
plt.plot(datenums,values)
plt.show()
```

# making value of 'longtime' column ready to plot in the graph

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as md
import numpy as np
import datetime as dt
import time
%matplotlib inline

start_index=650
end_index=800
interval_index=1#place it after end_index. start_index:end_index:interval_index
minute_interval=10

longTime=test.loc[start_index:end_index:interval_index,['longTime']]
longTime=longTime.as_matrix()
print(longTime.size)
a=[]
date_time_array=[]
for k in longTime:
    a=np.append(a,k)

timestamp_array=[]
count=0
for b in a[::-1]:
    b=int(b) # make plain integer
    str_b=str(b)
    c=str_b[-3:]
    new_str_b=str_b.replace(c, '',1)
    new_str_b_time=int(new_str_b)
    timestamp_array=np.append(timestamp_array,new_str_b_time)
# print(timestamp_array[::5])
duration=1000
# timestamp_arrays=[]
# timestamp_arrays=np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array))
# print(timestamp_arrays)
# print(np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array)))

date_list=[dt.datetime.fromtimestamp(ts) for ts in timestamp_array]
# for r in date_list:
#     print(r)
# print(date_list[::5])

datenums_array=md.date2num(date_list)
# for w in datenums_array:
#     print(w)
# print(type(datenums_array))

value_1=test.loc[start_index:end_index:interval_index,['RWWIHOA_MWT0','AEDAT_3_80V0']]
value_1=value_1.as_matrix()

RWWIHOA_MWT0=value_1[::,0]
RWWIHOA_MWT0=RWWIHOA_MWT0[::-1]

AEDAT_3_80V0=value_1[::,1]
AEDAT_3_80V0=AEDAT_3_80V0[::-1]

print(max(RWWIHOA_MWT0),'\n',min(RWWIHOA_MWT0))
print(max(AEDAT_3_80V0),'\n',min(AEDAT_3_80V0))

plt.title('graph')
plt.subplots_adjust(bottom=.2)
plt.xticks( rotation=25)
ax=plt.gca()

minlocator = md.MinuteLocator(interval=minute_interval) # to plot every minute of an range
# minlocator.MAXTICKS  = 40000
ax.xaxis.set_major_locator(minlocator)

xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
# plt.subplot(2,1,1)
plt.plot(datenums_array,RWWIHOA_MWT0,color='orange') #you can give here datenums_array
plt.plot(datenums_array,AEDAT_3_80V0,color='RED')
# plt.plot(datenums_array,value_1,color='red')
plt.gcf().autofmt_xdate() # to beautify the X label
# plt.title('2nd graph')
# plt.subplot(1,2,2) #1 row, 2 cols, position
# plt.xlabel('date_time')
# plt.ylabel('value_1_rev')
# plt.plot(datenums_array,value_1)
# plt.xlim((min(datenums_array),max(datenums_array)))
plt.xlim(datenums_array[0],datenums_array[-1])
# plt.ylim(40,60)
# print(min(datenums_array),max(datenums_array))
# print(datenums_array[0],datenums_array[-1])
plt.show()
```
