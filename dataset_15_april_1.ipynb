```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as md
import numpy as np
import datetime as dt
import time
%matplotlib inline

start_index=650
end_index=800
interval_index=1#place it after end_index. start_index:end_index:interval_index
minute_interval=10

longTime=test.loc[start_index:end_index:interval_index,['longTime']]
longTime=longTime.as_matrix()
print(longTime.size)
a=[]
date_time_array=[]
for k in longTime:
    a=np.append(a,k)

timestamp_array=[]
count=0
for b in a[::-1]:
    b=int(b) # make plain integer
    str_b=str(b)
    c=str_b[-3:]
    new_str_b=str_b.replace(c, '',1)
    new_str_b_time=int(new_str_b)
    timestamp_array=np.append(timestamp_array,new_str_b_time)
# print(timestamp_array[::5])
duration=1000
# timestamp_arrays=[]
# timestamp_arrays=np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array))
# print(timestamp_arrays)
# print(np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array)))

date_list=[dt.datetime.fromtimestamp(ts) for ts in timestamp_array]
# for r in date_list:
#     print(r)
# print(date_list[::5])

datenums_array=md.date2num(date_list)
# for w in datenums_array:
#     print(w)
# print(type(datenums_array))

value_1=test.loc[start_index:end_index:interval_index,['RWWIHOA_MWT0','AEDAT_3_80V0']]
value_1=value_1.as_matrix()

RWWIHOA_MWT0=value_1[::,0]
RWWIHOA_MWT0=RWWIHOA_MWT0[::-1]

AEDAT_3_80V0=value_1[::,1]
AEDAT_3_80V0=AEDAT_3_80V0[::-1]

print(max(RWWIHOA_MWT0),'\n',min(RWWIHOA_MWT0))
print(max(AEDAT_3_80V0),'\n',min(AEDAT_3_80V0))

plt.title('graph')
plt.subplots_adjust(bottom=.2)
plt.xticks( rotation=25)
ax=plt.gca()

minlocator = md.MinuteLocator(interval=minute_interval) # to plot every minute of an range
# minlocator.MAXTICKS  = 40000
ax.xaxis.set_major_locator(minlocator)

xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)


# plt.subplot(211)
plt.plot(datenums_array,RWWIHOA_MWT0,color='orange') #you can give here datenums_array
plt.title('Story of Power Plant')
plt.ylabel('RWWIHOA_MWT0')
plt.xlabel('Time')

# plt.subplot(212)
# plt.plot(datenums_array,AEDAT_3_80V0,color='RED')
# plt.xlabel('Time')
# plt.ylabel('AEDAT_3_80V0')
plt.gcf().autofmt_xdate() # to beautify the X label
# plt.xlim((min(datenums_array),max(datenums_array)))
plt.xlim(datenums_array[0],datenums_array[-1])
# plt.ylim(40,60)
# plt.subplots_adjust(hspace=1.4, wspace=1.4)
# plt.tight_layout() #to remove overlapping among of subplot
plt.show()
```

```python
start_index=650
end_index=800
interval_index=1#place it after end_index. start_index:end_index:interval_index
minute_interval=10

def time_graph(y_item):
    global start_index
    global end_index
    global interval_index
    global minute_interval
    
    longTime=test.loc[start_index:end_index:interval_index,['longTime']]
    longTime=longTime.as_matrix()
    print(longTime.size)
    a=[]
    date_time_array=[]
    for k in longTime:
        a=np.append(a,k)

    timestamp_array=[]
    count=0
    for b in a[::-1]:
        b=int(b) # make plain integer
        str_b=str(b)
        c=str_b[-3:]
        new_str_b=str_b.replace(c, '',1)
        new_str_b_time=int(new_str_b)
        timestamp_array=np.append(timestamp_array,new_str_b_time)
# print(timestamp_array[::5])
    duration=1000
# timestamp_arrays=[]
# timestamp_arrays=np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array))
# print(timestamp_arrays)
# print(np.linspace(timestamp_array[0],timestamp_array[0]+duration,len(timestamp_array)))

    date_list=[dt.datetime.fromtimestamp(ts) for ts in timestamp_array]
# for r in date_list:
#     print(r)
# print(date_list[::5])

    datenums_array=md.date2num(date_list)
    
    value_1=test.loc[start_index:end_index:interval_index,[y_item]]
    value_1=value_1.as_matrix()
    value_y=value_1[::,0]
    value_y=value_y[::-1]

    
    plt.title('graph')
    plt.subplots_adjust(bottom=.2)
    plt.xticks( rotation=25)
    ax=plt.gca()
    
    minlocator = md.MinuteLocator(interval=minute_interval) # to plot every minute of an range
# minlocator.MAXTICKS  = 40000
    ax.xaxis.set_major_locator(minlocator)
    
    xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
    ax.xaxis.set_major_formatter(xfmt)
    plt.plot(datenums_array,value_y,color='orange') #you can give here datenums_array
    plt.title('Story of Power Plant')
    plt.ylabel(y_item)
    plt.xlabel('Time')
    plt.gcf().autofmt_xdate() # to beautify the X label
# plt.xlim((min(datenums_array),max(datenums_array)))
    plt.xlim(datenums_array[0],datenums_array[-1])
# plt.ylim(40,60)
# plt.subplots_adjust(hspace=1.4, wspace=1.4)
# plt.tight_layout() #to remove overlapping among of subplot
    plt.show()


time_graph(y_item='RWWIHOA_MWT0')
time_graph(y_item='AEDAT_3_80V0')
```
