```python
import pandas as pd
import numpy as np
import time
import datetime
```

# Below csv file has generated from Sebastian Matzat - Dataset260419.table file

```python
test= pd.read_csv('E:/University of Bremen MSc/masters_thesis/IAT_sebastian/dataset_26_april_1.csv')
```

```python
longTime=test.loc[0:,['longTime']]
longTime=longTime.as_matrix()
a=[]
date_time_array=[]
for k in longTime:
    a=np.append(a,k)
str_time=[]
correct_longtime=[]
datetime_time=[]
count=0
for b in a:
    b=int(b) # make plain integer
    str_b=str(b)
    c=str_b[-3:]
    new_str_b=str_b.replace(c, '',1)
    new_str_b_time=int(new_str_b)
    correct_longtime.append(new_str_b_time)
    now_time=datetime.datetime.fromtimestamp(new_str_b_time)
    convert_time=now_time.strftime('%Y-%m-%d %H:%M')
    str_time.append(convert_time)
# print(str_time)
test_new = test.assign(stringTime=str_time,correct_longtime=correct_longtime) # here new column in the panda dataframe for string_time has added
```

```python
test_new['dateTime'] =  pd.to_datetime(test_new['stringTime'], format='%Y-%m-%d %H:%M')
```

```python
train_data=test_new.loc[0:8000,['dateTime','AEWIHOAVAB2','AEDATZ_HO_V0','RWWIHOA_HWT0', 'RWWIHOB_HWT0',
                                  'AEWIHOAKZ_V0', 'AEWIHOBKZ_V0','RWWIHOA_MWT0','RWWIHOB_MWT0',
                                  'AEWIHO_T9AV2','AEDAHO_T9_V2']]
```
